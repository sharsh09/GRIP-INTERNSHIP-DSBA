#Analysing the Data

```python
#number of seasons & matches played in every season.
num_matches_df = df.groupby('season')[['id']].count()
num_matches_df
```

```python
#Finding the number of teams who played IPL till 2019
print(df.team1.unique())
print('{} different teams played the IPL from 2008-2019'.format(df.team1.unique().shape[0]))
```

```python
#Number of matches won by each team from 2008-2019

winner_df = df.groupby('winner')[['id']].count()
winner_df=winner_df.sort_values('id', ascending=False).reset_index()

winner_df.rename(columns = { 'id': 'matches_won','winner':'team'}, inplace = True)
winner_df
```

```python
import plotly.express as px
# Create a pie chart for the number of matches won by each team
fig = px.pie(winner_df, names='team', values='matches_won', title='Number of Matches Won by Each Team (2008-2019)')

# Show the pie chart
fig.show()
```

```python
final_match_df = df.groupby('season').tail(1).sort_values('season').reset_index()
final_match_df
```

```python
#Count of Season Winners
print(final_match_df['winner'].value_counts())
plt.figure(figsize=(14, 6))
plt.title('Count of Season Winners',fontweight=800)
sns.countplot(x=final_match_df['winner'],palette=['#ECC5F2','#1C2C46','#FFFF48','#F0E1A1','#15244C','#D4480B']);
```

```python
#Count of Finalists in IPL
final_match_team = pd.concat([final_match_df['team1'],final_match_df['team2']])
final_match_team.value_counts()
```

```python
plt.title('Count of Finalists in IPL',fontweight=800)
plt.xlabel('Teams')
sns.countplot(x=final_match_team,palette=['#FFFF48','#1C2C46','#15244C','#CD202D','#D4480B','#ECC5F2','#EF2920','#F0E1A1','#E776CA'])
plt.ylabel('Final Matches Played')
plt.xticks(rotation=90);
```

```python
#Top 10 Player with Most Man of the Match Awards
mom_count =df.groupby('player_of_match')[['id']].count()
mom_count = mom_count.sort_values('id',ascending=False).head(10)
mom_count
```

```python
#Matches won by the teams while chasing and defending.

# Filter the DataFrame to include only matches where a result was achieved
df = df[df['result'] != 'no result']

# Calculate the total matches won by each team while chasing
chasing_wins = df[df['toss_decision'] == 'field'].groupby('winner')['id'].count().reset_index()
chasing_wins.rename(columns={'id': 'matches_won_chasing'}, inplace=True)

# Calculate the total matches won by each team while defending
defending_wins = df[df['toss_decision'] == 'bat'].groupby('winner')['id'].count().reset_index()
defending_wins.rename(columns={'id': 'matches_won_defending'}, inplace=True)

# Merge the two DataFrames to get the total wins for each team
total_wins = pd.merge(chasing_wins, defending_wins, on='winner', how='outer').fillna(0)

# Calculate the overall total wins for each team
total_wins['total_wins'] = total_wins['matches_won_chasing'] + total_wins['matches_won_defending']

# Display the total wins for each team
print(total_wins)
```

```python
# Plot the results as a bar chart
plt.figure(figsize=(12, 6))
plt.bar(total_wins['winner'], total_wins['matches_won_chasing'], label='Chasing Wins', color='skyblue')
plt.bar(total_wins['winner'], total_wins['matches_won_defending'], label='Defending Wins', color='lightgreen', bottom=total_wins['matches_won_chasing'])
plt.xlabel('Teams')
plt.ylabel('Number of Matches Won')
plt.title('Matches Won by Teams While Chasing and Defending')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Rate of winning the toss as well as the match.

# Create a DataFrame to store team-wise statistics
team_stats = pd.DataFrame()

# Calculate the number of matches played by each team
team_stats['matches_played'] = df.groupby('team1')['id'].count() + df.groupby('team2')['id'].count()

# Calculate the number of tosses won by each team
team_stats['toss_won'] = df[df['toss_winner'] == df['team1']].groupby('toss_winner')['id'].count() + df[df['toss_winner'] == df['team2']].groupby('toss_winner')['id'].count()

# Calculate the number of matches won by each team
team_stats['matches_won'] = df[df['winner'] == df['team1']].groupby('winner')['id'].count() + df[df['winner'] == df['team2']].groupby('winner')['id'].count()

# Calculate the rate of winning the toss and winning the match
team_stats['winning_rate'] = (team_stats['matches_won'] / team_stats['toss_won']).fillna(0)

# Reset the index to have 'team' as a column instead of the index
team_stats.reset_index(inplace=True)

# Display the team-wise statistics
print(team_stats)
```

```python
# Filter the DataFrame to include only matches with a result
df = df[df['result'] != 'no result']

# Sort the DataFrame by matches won by runs in descending order
top_runs_wins = df.sort_values(by='win_by_runs', ascending=False).head(10)

# Sort the DataFrame by matches won by wickets in descending order
top_wickets_wins = df.sort_values(by='win_by_wickets', ascending=False).head(10)

# Create subplots for both top 10 runs and wickets wins
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot the top 10 teams with the most runs wins
axes[0].barh(top_runs_wins['winner'], top_runs_wins['win_by_runs'], color='skyblue')
axes[0].set_xlabel('Runs Won By')
axes[0].set_title('Top 10 Teams - Most Runs Wins')

# Plot the top 10 teams with the most wickets wins
axes[1].barh(top_wickets_wins['winner'], top_wickets_wins['win_by_wickets'], color='lightcoral')
axes[1].set_xlabel('Wickets Won By')
axes[1].set_title('Top 10 Teams - Most Wickets Wins')

plt.tight_layout()
plt.show()
```

```python
#Top 10 Cities with the maximum number of matches played
city_counts= df.groupby('city').apply(lambda x:x['city'].count()).reset_index(name='Match Counts')
top_cities_order=city_counts.sort_values(by='Match Counts',ascending=False)
top_cities=top_cities_order[:10]
print('Top 10 Cities with the maximum number of matches played:\n',top_cities)
plt.figure(figsize=(7,7))
plt.pie(top_cities['Match Counts'],labels=top_cities['city'],autopct='%1.1f%%', startangle=30)
plt.axis('equal')
plt.title('Top Cities that have hosted IPL Matches',size=10)
```
